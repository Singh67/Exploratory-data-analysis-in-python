```python
from matplotlib import rcParams
import plotly.tools as tls
import plotly.offline as py
from plotly.offline import init_notebook_mode,iplot,plot
from plotly.graph_objs import *
init_notebook_mode(connected=True)
import warnings 
```

```python
import numpy as np
import pandas as pd
```

```python
emissions=pd.read_csv('Co2EmissionsPerCapitaByCountry.csv',index_col=0)
```

```python
emissions.head()
```

```python
sumcolumnSelection=['United Arab Emirates | ARE','United Kingdom | GBR','United States | USA',
                  'China | CHN','India | IND','Year']
stackedAreaData=emissions.loc[:,(sumcolumnSelection)]
```

```python
stackedAreaData.head()
```

```python
stackedAreaData['Total']=stackedAreaData[sumcolumnSelection].sum(axis=1)
```

```python
stackedAreaData.head()
```

```python
for country in sumcolumnSelection:
    stackedAreaData['pc_' + str(country)]=stackedAreaData[country]/stackedAreaData['Total']
```

```python
stackedAreaData.head()
```

```python
sumcolumnSelectionPC=['pc_United Arab Emirates | ARE','pc_United Kingdom | GBR','pc_United States | USA',
                  'pc_China | CHN','pc_India | IND']
```

```python
PCAreaData=stackedAreaData[sumcolumnSelectionPC].cumsum(axis=1)

PCAreaData.head()
```

```python
PCAreaData['Year']=stackedAreaData['Year']
```

```python
PCAreaData.head()
```

```python
traces=[]

for col in PCAreaData.columns.tolist():
    if col!='Year':
        traces.append({'type':'scatter',
                       'x':PCAreaData['Year'],
                       'y':PCAreaData[col],
                       'name': col[3:-6],
                       'mode': 'lines',
                       'fill':'tonexty'
                      })

data=Data(traces)
layout={'title':'Proportion of CO2 Emission , 1960-2011',
        'xaxis':{'title':'Year'},
        'yaxis':{'title':'Proportion of co2 Emissions'}}

fig=Figure(data=data,layout=layout)
iplot(fig)
```
