```python
from matplotlib import rcParams
import plotly.tools as tls
import plotly.offline as py
from plotly.offline import init_notebook_mode,iplot,plot
from plotly.graph_objs import *
init_notebook_mode(connected=True)
import warnings 
```

```python
import numpy as np
import pandas as pd
```

```python
companies=pd.read_csv('CompaniesRevenueEmployees.csv',encoding = 'unicode_escape',index_col=0)
```

```python
companies['Text']=companies.apply(lambda x: "<b>{}</b><i>{}</i><br>Revenue:${} billion<br> Employee{:,}".format(x['Name'],
                                  x['Industry'],x['Revenue (USD billions)'],x['Employees']),axis=1)
```

```python
companies.loc[10,'Text']
```

```python
companiesTrace={'type':'scatter',
                'mode':'markers',
                'x':companies['Revenue (USD billions)'],
                'y':companies['Employees'],
                'text': companies['Text'],
                'hoverinfo':'text'}

layout={'title':'Revenue by number of employees',
        'xaxis':{'title':'Revenue (Billion $)'},
        'yaxis':{'title':'Employee'},
        'hovermode':'closest'}

```

```python
data=Data([companiesTrace])
fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
#Styling the marker points
```

```python
companiesTrace.update({'marker' :{'symbol':'square-open'}})
data=Data([companiesTrace])
fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
companiesTrace.update({'marker' :{'opacity':0.7}})
data=Data([companiesTrace])
fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
companiesTrace.update({'marker' :{'color':'#6a5acd'}})
data=Data([companiesTrace])
fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
companiesTrace.update({'marker' :{'line':{'width':1.5,'color':'black'},'size':10}})
data=Data([companiesTrace])
fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
iris=pd.read_csv('irisDataset.csv',index_col=0)
```

```python
iris.head()
```

```python
irisSpeciesUnique=list(iris['Species'].unique())
irisSpeciesUnique
```

```python
traces=[]

for species in irisSpeciesUnique:
    traces.append({'type':'scatter',
                   'mode':'markers',
                   'x':iris.loc[iris['Species']==species,'Sepal length'],
                   'y':iris.loc[iris['Species']==species,'Sepal width'],
                   'name': species})
```

```python
layout={'title':'Sepal length and width for iris setosa, versicolor,virginica',
        'xaxis':{'title':'Sepal lenght(cm)',
                  'range':[iris['Sepal length'].min() *.95,iris['Sepal length'].max() *1.05]},
        'yaxis':{'title': 'Sepal width',
                'range':[iris['Sepal width'].min() *.95,iris['Sepal width'].max() *1.05]}}
                        
fig=Figure(data=traces,layout=layout)
```

```python
iplot(fig)
```
