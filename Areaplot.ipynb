```python
from matplotlib import rcParams
import plotly.tools as tls
import plotly.offline as py
from plotly.offline import init_notebook_mode,iplot,plot
from plotly.graph_objs import *
init_notebook_mode(connected=True)
import warnings 
```

```python
import numpy as np
import pandas as pd
```

```python
emissions=pd.read_csv('Co2EmissionsPerCapitaByCountry.csv',index_col=0)
```

```python
emissions.head()
```

```python
columnNames=emissions.columns.tolist()
```

```python
UkEmissions={'type':'scatter',
             'x':emissions['Year'],
             'y':emissions['United Kingdom | GBR'],
             'mode':'lines',
             'name': 'UK C02 Emission'
            }
layout={'title':'C02 emission in kilons for UK, 1960-2015',
        'xaxis':{'title':'Year'},
        'yaxis':{'title':'Co2 Emission (kt)'}
       }
data=Data([UkEmissions])

fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
UkEmissions.update({'fill':'tozeroy'})
data=Data([UkEmissions])
fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
UkEmissions.update({'fillcolor':'rgba(89,100,212,0.46)'})
data=Data([UkEmissions])
fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
UkEmissions.update({'fill':'tozerox'})
data=Data([UkEmissions])
fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
UkEmissions.update({'x':emissions['United Kingdom | GBR'],
                    'y':emissions['Year']})
layout={'title':'C02 emission in kilons for UK, 1960-2015',
        'xaxis':{'title':'Co2 Emission (kt)'},
        'yaxis':{'title':'Year'}
       }
data=Data([UkEmissions])

fig=Figure(data=data,layout=layout)
iplot(fig)
```

```python
#Stacked Area chart
```

```python
columnSelection=['United Arab Emirates | ARE','United Kingdom | GBR','United States | USA',
                  'China | CHN','India | IND','Year']

stackedAreaData=emissions.loc[:,(columnSelection)]

stackedAreaData.head()
```

```python
UnitedArabEmirates={'type':'scatter',
             'x':stackedAreaData['Year'],
             'y':stackedAreaData['United Arab Emirates | ARE'],
             'mode':'lines',
             'fill':'tonexty',
             'name': 'UAE C02 Emission'
            }

UnitedKingdom ={'type':'scatter',
             'x':stackedAreaData['Year'],
             'y':stackedAreaData['United Arab Emirates | ARE'] + stackedAreaData['United Kingdom | GBR'],
             'mode':'lines',
             'fill':'tonexty',
             'name': 'GBR C02 Emission'
            }


UnitedStates ={'type':'scatter',
             'x':stackedAreaData['Year'],
             'y':stackedAreaData['United Arab Emirates | ARE'] + stackedAreaData['United Kingdom | GBR'] +
                  stackedAreaData['United States | USA'],
             'mode':'lines',
             'fill':'tonexty',
             'name': 'USA C02 Emission'
            }

China ={'type':'scatter',
             'x':stackedAreaData['Year'],
             'y':stackedAreaData['United Arab Emirates | ARE'] + stackedAreaData['United Kingdom | GBR'] +
                  stackedAreaData['United States | USA'] + stackedAreaData['China | CHN'],
             'mode':'lines',
             'fill':'tonexty',
             'name': 'China C02 Emission'
            }

India={'type':'scatter',
             'x':stackedAreaData['Year'],
             'y':stackedAreaData['United Arab Emirates | ARE'] + stackedAreaData['United Kingdom | GBR'] +
                  stackedAreaData['United States | USA'] + stackedAreaData['China | CHN'] +
                  stackedAreaData['India | IND'],
             'mode':'lines',
             'fill':'tonexty',
             'name': 'India C02 Emission'
            }
data=Data([UnitedArabEmirates,UnitedKingdom,UnitedStates,China,India])
```

```python
layout={'title':'C02 emission in kilons, 1960-2015',
        'xaxis':{'title':'Year'},
        'yaxis':{'title':'Co2 Emission (kt)'}
       }
```

```python
fig=Figure(data=data, layout=layout)
```

```python
iplot(fig)
```
